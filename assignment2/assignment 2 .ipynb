## 2. Sort by the number of comments / by the price
By this way, I found that the number of comments doesn't related with the price of phone in some degree. It means that more comments doesn't equal to the higher price.
In the first chart, we can see that the highest discussion is the price of 3888. In the third chart, people are more focus on phones with prices above 1000 and below 5000.

```python
int_comm=[]
for i in data['comment']:
    a = i.replace('+','')
    int_comm.append(int(a))
data['comment']=int_comm
comment_a=data.sort_values(by='comment', ascending=False)[:20]
comment_a
```

```python
int_price=[]
for i in data['price']:
    int_price.append(int(i))
data['price']=int_price
price_a=data.sort_values(by='price', ascending=False)[:20]
price_a
```

## 3. Shop with high price or high comment
Through the two charts on the right, I found that the price of phones in '华为京东自营官方旗舰店' are higher than other shop, maybe they provide many types of phones and some of then are high-quality and high-tech. And the amout of comments in '荣耀京东自营旗舰店' are higher than '华为京东自营官方旗舰店', so I assume that '荣耀京东自营旗舰店' is more popular than '华为京东自营官方旗舰店'.

```python
fig, axes = plt.subplots(nrows=2, ncols=2,constrained_layout=True,figsize=(30,20)) 

price_comment = price_a[['comment','price']]
ax1 = price_comment.plot(kind = 'barh',x='price',y='comment',color = '#87CEEB',ax=axes[0,0],fontsize=24)
ax1.set_title("top 20 price vs comment",fontsize=40)

price_shop = price_a[['price','shop']]
ax2 = price_shop.plot(kind = 'bar',x='shop',y='price',color = '#46bc99',ax=axes[0,1],fontsize=24)
ax2.set_title("shop with price amount",fontsize=40)

comment_price = comment_a[['price','comment']]
ax3 = comment_price.plot(kind = 'barh',x='comment',y='price',color = '#87CEEB',ax=axes[1,0],fontsize=24)
ax3.set_title("top 20 comment vs price",fontsize=40)

comment_shop = comment_a[['comment','shop']]
ax4 = comment_shop.plot(kind = 'bar',x='shop',y='comment',color = '#46bc99',ax=axes[1,1],fontsize=24)
ax4.set_title("shop with comment amount",fontsize=40)
```
