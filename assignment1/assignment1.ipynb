```python
from selenium import webdriver
import time

url='https://list.jd.com/list.html?cat=9987,653,655&ev=exbrand_8557'
browser = webdriver.Chrome()
browser.get(url)
```

```python
all_phone=[]

for i in range(0,4):
    page_url='{url_initial}&page={number}&sort=sort_rank_asc&trans=1&JL=6_0_0#J_main'.format(url_initial = url,number=i)
    browser.get(page_url)
    huawei_phones = browser.find_elements_by_class_name("gl-item")
        
        
    for one_phone in huawei_phones:
        huawei_data={}
        huawei_data['name']=one_phone.find_element_by_xpath("//div[@class='p-name']").find_element_by_tag_name('a').text
        huawei_data['price']=one_phone.find_element_by_class_name('p-price').find_element_by_tag_name('i').text
        huawei_data['shop']=one_phone.find_element_by_class_name('p-shop').find_element_by_tag_name('a').text
        huawei_data['comment']=one_phone.find_element_by_css_selector('.goods-list-v2 .gl-item .p-commit strong a').text
        huawei_data['type']=one_phone.find_element_by_xpath("//div[@class='p-icons J-pro-icons']").find_element_by_tag_name('i').text
        all_phone.append(huawei_data)
       
```

```python
all_phone
```

```python
import pandas as pd 
df = pd.DataFrame(all_phone)
df
```

```python
df['comment']=df['comment'].str.replace('万','00000')
df['comment']=df['comment'].str.replace('.','')
df
```

```python
import csv
with open('huawei_data.csv', 'w') as f:
    my_writer = csv.writer(f)
    header = ['name', 'price','shop','comment','type']
    my_writer.writerow(header)
    my_writer.writerows(all_phone)
```
